```python
import pandas as pd

df = pd.read_csv("marketing_campaign.csv", sep="\t")


print(df.head())
print(df.info())
print(df.isnull().sum())
```

```python
df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]
df.columns
```

```python
df = df.drop_duplicates()

```

```python
df['income'] = df['income'].fillna(df['income'].median())
```

```python
df['education'] = df['education'].str.lower().str.strip()
df['marital_status'] = df['marital_status'].str.lower().str.strip()
```

```python
print(df['education'].unique())
print(df['marital_status'].unique())
```

```python
df['marital_status'] = df['marital_status'].replace({
    'absurd': 'single',
    'yolo': 'single'
})
```

```python
df['dt_customer'] = pd.to_datetime(df['dt_customer'], dayfirst=True)
```

```python
df['dt_customer_dd_mm_yyyy'] = df['dt_customer'].dt.strftime('%d-%m-%Y')
```

```python
df['income'] = df['income'].astype(int)
```

```python
q1 = df['income'].quantile(0.25)
q3 = df['income'].quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

df['income_outlier'] = (df['income'] < lower) | (df['income'] > upper)
```

```python
print(df.info())
print(df.isnull().sum())
print(df.head())
```

```python
df.to_csv("customer_personality_cleaned.csv", index=False)

```
